```python
# Funzione per calcolare la moda
def mode(series):
    return series.mode().iloc[0]

agg_dict = {
    # ---------- FEATURE NUMERICHE → MEDIA ----------
    "popularity": "mean",
    "duration_ms": "mean",
    "danceability": "mean",
    "energy": "mean",
    "loudness": "mean",
    "speechiness": "mean",
    "acousticness": "mean",
    "instrumentalness": "mean",
    "liveness": "mean",
    "valence": "mean",
    "tempo": "mean",

    # ---------- FEATURE CATEGORICHE → MODA ----------
    "album_name": mode,
    "track_genre": mode,
    "explicit": mode,
    "key": mode,
    "mode": mode,
    "time_signature": mode,

    # ---------- METADATO ----------
    "track_id": "first"
}


# Deduplicazione semantica avanzata

df_dedup = (
    df
    .groupby(["track_name", "artists"], as_index=False)
    .agg(agg_dict)
)



# CONTROLLI FINALI

# 1. Nessun duplicato semantico residuo
assert df_dedup.duplicated(
    subset=["track_name", "artists"]
).sum() == 0, "Errore: duplicati semantici presenti!"

# 2. Reset indice
df_dedup = df_dedup.reset_index(drop=True)

# 3. Info finale
print("Dataset finale pronto per il recommender")
print("Shape:", df_dedup.shape)
df_dedup.info()

df=df_dedup
```

Il dataset contiene più versioni della stessa canzone (album diversi, ristampe, compilation), identificate dalla coppia (track_name, artists). Per ottenere una rappresentazione unica e robusta di ciascun brano, le versioni duplicate vengono aggregate mantenendo una sola riga:
 - le feature numeriche sono aggregate tramite media
 - le feature categoriche tramite moda
  Questo approccio riduce il rumore e fornisce una rappresentazione più stabile del contenuto musicale per il sistema di raccomandazione.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selezione delle audio features
audio_features = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'speechiness',
    'loudness',
    'tempo',
    'duration_ms'
]

# Matrice di correlazione
corr = df[audio_features].corr(method='pearson')

# Visualizzazione
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    square=True
)
plt.title("Matrice di correlazione delle audio features")
plt.show()
```


La matrice di correlazione mostra alcune relazioni significative tra le audio features. In particolare, energy è fortemente correlata con loudness e negativamente con acousticness, confermando che brani più energetici tendono a essere meno acustici e più rumorosi. Danceability e valence presentano una correlazione moderata, utile per modellare il mood musicale. Feature come instrumentalness, speechiness e tempo risultano debolmente correlate con le altre, indicando che forniscono informazione complementare. Nessuna coppia di feature è perfettamente correlata, quindi le feature selezionate sono adatte a essere utilizzate insieme in un sistema di raccomandazione basato sul contenuto. La feature duration_ms mostra correlazioni trascurabili ed è quindi esclusa dal calcolo della similarità.


```python
plt.figure(figsize=(6,5))
plt.scatter(df['energy'], df['loudness'], alpha=0.3)
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energy vs Loudness')
plt.show()

#Brani più energetici tendono ad avere maggiore loudness.(corr 0.76)
```

```python
# Suddividiamo energy in quartili
df['energy_bin'] = pd.qcut(df['energy'], 4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])

plt.figure(figsize=(7,5))
for level in df['energy_bin'].cat.categories:
    subset = df[df['energy_bin'] == level]
    plt.hist(subset['loudness'], bins=30, alpha=0.5, label=level)

plt.xlabel('Loudness')
plt.ylabel('Count')
plt.title('Distribuzione di Loudness per livelli di Energy')
plt.legend(title='Energy level')
plt.show()
```

Il grafico mostra che all’aumentare del livello di energy cresce anche il valore medio di loudness. Questo conferma una forte relazione positiva tra le due feature, già osservata nella matrice di correlazione (ρ ≈ 0.76), indicando che i brani più energetici tendono a essere anche più rumorosi.

```python
# Suddividiamo duration in quartili
df['duration_bin'] = pd.qcut(
    df['duration_ms'],
    4,
    labels=['Short', 'Medium-Short', 'Medium-Long', 'Long']
)

plt.figure(figsize=(7,5))
for level in df['duration_bin'].cat.categories:
    subset = df[df['duration_bin'] == level]
    plt.hist(subset['energy'], bins=30, alpha=0.5, label=level)

plt.xlabel('Energy')
plt.ylabel('Count')
plt.title('Distribuzione di Energy per durata del brano')
plt.legend(title='Duration')
plt.show()
```

FEATURE ENGINEERING

```python
import matplotlib.pyplot as plt

# Mood score: combina energia ed emozione
df['mood_score'] = (df['energy'] + df['valence']) / 2

# Dance mood: musica ballabile e positiva
df['dance_mood'] = (df['danceability'] + df['valence']) / 2

# Electronic index: contrapposizione tra elettronico e acustico
df['electronic_index'] = df['energy'] - df['acousticness']

# Flag strumentale
df['is_instrumental'] = (df['instrumentalness'] > 0.5).astype(int)

#Grafici 
# Mood Score 
plt.figure(figsize=(6,4))
plt.hist(df['mood_score'], bins=30)
plt.xlabel('Mood score')
plt.ylabel('Count')
plt.title('Distribuzione del Mood Score')
plt.show()


# Electronic Index per genere 
plt.figure(figsize=(8,4))
df.boxplot(column='electronic_index', by='main_genre', rot=45)
plt.xlabel('Main genre')
plt.ylabel('Electronic index')
plt.title('Electronic Index per genere principale')
plt.suptitle('')
plt.show()
```

Sono state introdotte alcune feature derivate per rappresentare aspetti musicali di livello più alto, come il mood complessivo e la componente elettronica del brano. I grafici mostrano che tali feature risultano ben distribuite e musicalmente interpretabili, supportando il loro utilizzo nel sistema di raccomandazione, in particolare nella fase di cold start.


Scaling finale con MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Selezione features finali
base_features = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'speechiness',
    'loudness',
    'tempo'
]

engineered_features = [
    'mood_score',
    'electronic_index',
    'is_instrumental'
]

final_features = base_features + engineered_features



# SCALING
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df[final_features])

# DataFrame scalato (comodo per debug / analisi)
df_scaled = pd.DataFrame(
    X_scaled,
    columns=final_features,
    index=df.index
)

display(df_scaled.head())
```

Dopo la selezione delle feature originali e derivate, è stato applicato uno scaling Min-Max per riportare tutte le variabili nello stesso intervallo [0,1]. Questo passaggio è fondamentale per evitare che feature con scale diverse influenzino in modo sproporzionato il calcolo della similarità tra brani.
