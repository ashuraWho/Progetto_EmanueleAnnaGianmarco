```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv')
display(df.head())

# Info dataset
display(df.info())

# Statistiche descrittive
display(df.describe())
```

```python
# Valori mancanti
display(df.isna().sum().sort_values(ascending=False))

# Duplicati
display(df.duplicated().sum())
```

```python
print("Righe prima:", len(df))
df = df.dropna()
print("Righe dopo:", len(df))
```

```python
df = (
    df.dropna()
      .groupby(['track_name', 'artists', 'album_name'], as_index=False)
      .agg({
          **{c: 'mean' for c in df.select_dtypes(include='number').columns},
          **{c: 'first' for c in df.select_dtypes(exclude='number').columns
             if c not in ['track_name', 'artists', 'album_name']}
      })
)

# Info dataset
display(df.info())
```

```python
songs_per_genre = df['track_genre'].value_counts(ascending=True)

display(songs_per_genre)

import matplotlib.pyplot as plt

songs_per_genre.plot(kind='bar', figsize=(12,4))
plt.ylabel('Numero di canzoni')
plt.title('Numero di canzoni per genere')
plt.show()
```
