```python
df = (
    df.dropna()
      .groupby(['track_name', 'artists', 'album_name'], as_index=False)
      .agg({
          **{c: 'mean' for c in df.select_dtypes(include='number').columns},
          **{c: 'first' for c in df.select_dtypes(exclude='number').columns
             if c not in ['track_name', 'artists', 'album_name']}
      })
)

# Info dataset
display(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selezione delle audio features
audio_features = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'speechiness',
    'loudness',
    'tempo',
    'duration_ms'
]

# Matrice di correlazione
corr = df[audio_features].corr(method='pearson')

# Visualizzazione
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    square=True
)
plt.title("Matrice di correlazione delle audio features")
plt.show()
```


La matrice di correlazione mostra alcune relazioni significative tra le audio features. In particolare, energy Ã¨ fortemente correlata con loudness e negativamente con acousticness, confermando che brani piÃ¹ energetici tendono a essere meno acustici e piÃ¹ rumorosi. Danceability e valence presentano una correlazione moderata, utile per modellare il mood musicale. Feature come instrumentalness, speechiness e tempo risultano debolmente correlate con le altre, indicando che forniscono informazione complementare. Nessuna coppia di feature Ã¨ perfettamente correlata, quindi le feature selezionate sono adatte a essere utilizzate insieme in un sistema di raccomandazione basato sul contenuto. La feature duration_ms mostra correlazioni trascurabili ed Ã¨ quindi esclusa dal calcolo della similaritÃ .


```python
plt.figure(figsize=(6,5))
plt.scatter(df['energy'], df['loudness'], alpha=0.3)
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energy vs Loudness')
plt.show()

#Brani piÃ¹ energetici tendono ad avere maggiore loudness.(corr 0.76)
```

```python
# Suddividiamo energy in quartili
df['energy_bin'] = pd.qcut(df['energy'], 4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])

plt.figure(figsize=(7,5))
for level in df['energy_bin'].cat.categories:
    subset = df[df['energy_bin'] == level]
    plt.hist(subset['loudness'], bins=30, alpha=0.5, label=level)

plt.xlabel('Loudness')
plt.ylabel('Count')
plt.title('Distribuzione di Loudness per livelli di Energy')
plt.legend(title='Energy level')
plt.show()
```
