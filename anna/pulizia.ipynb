```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selezione delle audio features
audio_features = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'speechiness',
    'loudness',
    'tempo',
    'duration_ms'
]

# Matrice di correlazione
corr = df[audio_features].corr(method='pearson')

# Visualizzazione
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    square=True
)
plt.title("Matrice di correlazione delle audio features")
plt.show()
```


La matrice di correlazione mostra alcune relazioni significative tra le audio features. In particolare, energy Ã¨ fortemente correlata con loudness e negativamente con acousticness, confermando che brani piÃ¹ energetici tendono a essere meno acustici e piÃ¹ rumorosi. Danceability e valence presentano una correlazione moderata, utile per modellare il mood musicale. Feature come instrumentalness, speechiness e tempo risultano debolmente correlate con le altre, indicando che forniscono informazione complementare. Nessuna coppia di feature Ã¨ perfettamente correlata, quindi le feature selezionate sono adatte a essere utilizzate insieme in un sistema di raccomandazione basato sul contenuto. La feature duration_ms mostra correlazioni trascurabili ed Ã¨ quindi esclusa dal calcolo della similaritÃ .

Il grafico mostra che allâ€™aumentare del livello di energy cresce anche il valore medio di loudness.
```python
# Suddividiamo duration in quartili
df['duration_bin'] = pd.qcut(
    df['duration_ms'],
    4,
    labels=['Short', 'Medium-Short', 'Medium-Long', 'Long']
)

plt.figure(figsize=(7,5))
for level in df['duration_bin'].cat.categories:
    subset = df[df['duration_bin'] == level]
    plt.hist(subset['energy'], bins=30, alpha=0.5, label=level)

plt.xlabel('Energy')
plt.ylabel('Count')
plt.title('Distribuzione di Energy per durata del brano')
plt.legend(title='Duration')
plt.show()
```

FEATURE ENGINEERING

```python
import matplotlib.pyplot as plt

# Mood score: combina energia ed emozione
df['mood_score'] = (df['energy'] + df['valence']) / 2

# Dance mood: musica ballabile e positiva
df['dance_mood'] = (df['danceability'] + df['valence']) / 2

# Electronic index: contrapposizione tra elettronico e acustico
df['electronic_index'] = df['energy'] - df['acousticness']

# Flag strumentale
df['is_instrumental'] = (df['instrumentalness'] > 0.5).astype(int)

#Grafici 
# Mood Score 
plt.figure(figsize=(6,4))
plt.hist(df['mood_score'], bins=30)
plt.xlabel('Mood score')
plt.ylabel('Count')
plt.title('Distribuzione del Mood Score')
plt.show()


# Electronic Index per genere 
plt.figure(figsize=(8,4))
df.boxplot(column='electronic_index', by='main_genre', rot=45)
plt.xlabel('Main genre')
plt.ylabel('Electronic index')
plt.title('Electronic Index per genere principale')
plt.suptitle('')
plt.show()
```

Sono state introdotte alcune feature derivate per rappresentare aspetti musicali di livello piÃ¹ alto, come il mood complessivo e la componente elettronica del brano. I grafici mostrano che tali feature risultano ben distribuite e musicalmente interpretabili, supportando il loro utilizzo nel sistema di raccomandazione, in particolare nella fase di cold start.


Scaling finale con MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Selezione features finali
base_features = [
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'instrumentalness',
    'speechiness',
    'loudness',
    'tempo'
]

engineered_features = [
    'mood_score',
    'electronic_index',
    'is_instrumental'
]

final_features = base_features + engineered_features



# SCALING
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df[final_features])

# DataFrame scalato (comodo per debug / analisi)
df_scaled = pd.DataFrame(
    X_scaled,
    columns=final_features,
    index=df.index
)

display(df_scaled.head())
```

Dopo la selezione delle feature originali e derivate, Ã¨ stato applicato uno scaling Min-Max per riportare tutte le variabili nello stesso intervallo [0,1]. Questo passaggio Ã¨ fondamentale per evitare che feature con scale diverse influenzino in modo sproporzionato il calcolo della similaritÃ  tra brani.
